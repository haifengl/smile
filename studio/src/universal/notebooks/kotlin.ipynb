```kotlin
val home = System.getProperty("smile.home")
val iris = Read.arff(home + "/data/weka/iris.arff")
println(iris)
iris.summary()
```

```kotlin
val rf = RandomForest.fit(Formula.lhs("class"), iris)
println("OOB error = ${rf.metrics()}")
```
